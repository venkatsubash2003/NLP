```python
import pandas as pd
```

```python
df = pd.read_csv("all_kindle_review.csv")
```

```python
df["rating"] = df["rating"].apply(lambda x: 1 if x >= 3 else 0)
```

```python
df.head()
```

```python
from nltk.tokenize import sent_tokenize
```

```python
df["rating"].unique()
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
from nltk.corpus import stopwords
```

```python
import re
```

```python
import nltk
nltk.download("stopwords")
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
nltk.download("punkt")
```

```python
corpus = []
for i in range(len(df)):
  review = re.sub('[^a-zA-Z0-9]',' ',df["reviewText"][i])
  review = review.lower()
  review = review.split()
  review = [lemmatizer.lemmatize(word) for word in review if word not in stopwords.words("english")]
  review = " ".join(review)
  corpus.append(review)
```

```python
nltk.download("wordnet")
```

```python
corpus
```

```python
!pip install gensim
```

```python
from gensim.models import Word2Vec, KeyedVectors
```

```python
from gensim.utils import simple_preprocess
```

```python
words = []
for sent in corpus:
  sent_token = sent_tokenize(sent)
  for sentence in sent_token:
    words.append(simple_preprocess(sentence))
```

```python
words
```

```python
model = Word2Vec(words)
```

```python
model.wv.index_to_key
```

```python
model.corpus_count
```

```python
model.epochs
```

```python
words[0]
```

```python
import numpy as np
```

```python
def avg_word2vec(doc):
  return np.mean([model.wv[word] for word in doc if word in model.wv.index_to_key],axis=0) ## calculating mean horizontally or in horizontal direction
```

```python
!pip install tqdm
```

```python
from tqdm import tqdm
```

```python
X = []
for i in tqdm(range(len(words))):
  X.append(avg_word2vec(words[i]))
```

```python
X
```

```python
len(X)
```

```python
X[0].shape
```

```python
y = pd.get_dummies(df["rating"])
y = y.iloc[:,0].values
```

```python
y
```

```python
X= np.array(X)
```

```python
X = pd.DataFrame(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier()
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
X_train,X_test,y_train,y_test = train_test_split(df["reviewText"],df["rating"],test_size=0.2,random_state=42)
```

```python
clf = MultinomialNB()
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```
