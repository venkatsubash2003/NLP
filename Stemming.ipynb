```python
import nltk
nltk.download("punkt")
```

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
```

```python
words = ["Eating","eaten","eat","gone","going","go","gone","programming","program"]
for word in words:
  print(word + "-->" + stemmer.stem(word))
```

```python
stemmer.stem("congratulations")
```

```python
stemmer.stem("sitting")
```

```python
from nltk.stem import RegexpStemmer
stemmer = RegexpStemmer("ing$|e$|ed$|able$|")
```

```python
stemmer.stem("cars")
```

```python
stemmer.stem("going")
```

```python
stemmer.stem("eating")
```

```python
stemmer.stem("capable")
```
